```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Zomato dataset
df = pd.read_csv('/content/zomato.csv', on_bad_lines='skip', engine='python')


# Display the first few rows
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['rate'], kde=True, bins=20)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count occurrences of each cuisine and visualize the top 10
cuisine_counts = df['cuisines'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(y=cuisine_counts.index, x=cuisine_counts.values, palette='viridis')
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['approx_cost(for two people)'], kde=True, bins=20)
plt.title('Cost Distribution for Two People')
plt.xlabel('Cost (INR)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Online ordering analysis
online_order_counts = df['online_order'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=online_order_counts.index, y=online_order_counts.values, palette='coolwarm')
plt.title('Restaurants Offering Online Orders')
plt.xlabel('Online Order Available')
plt.ylabel('Number of Restaurants')
plt.show()

# Table booking analysis
book_table_counts = df['book_table'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=book_table_counts.index, y=book_table_counts.values, palette='coolwarm')
plt.title('Restaurants Offering Table Booking')
plt.xlabel('Table Booking Available')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='approx_cost(for two people)', y='rate', alpha=0.6)
plt.title('Cost vs Rating')
plt.xlabel('Cost for Two People (INR)')
plt.ylabel('Rating')
plt.show()
```
